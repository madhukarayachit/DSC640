```python
#Load in libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
population = pd.read_excel("Data/world-population.xls")
population.head()
```

```python
#Line chart using matplotlib
plt.plot(population['Year'], population['Population'], color = 'red')
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.title('World Population by Year')
plt.grid(True)
plt.show()
```

```python
#Step chart using world population data.
x = population['Year']
y = population['Population']

plt.step(x,y)
plt.xlabel('Year')
plt.ylabel('Population in Billions')
plt.title('World Population by Year')
plt.grid(True)
plt.show()
```
